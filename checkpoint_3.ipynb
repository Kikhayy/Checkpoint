```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("/content/supermarket dataset.zip")

df
```

```
plt.figure(figsize=(8,5))
sns.countplot(x="Branch", data=df, hue="Branch", palette="Blues_r", legend=False)
plt.title("Distribusi Penjualan per Cabang Supermarket", fontweight="bold")
plt.xlabel("Cabang")
plt.ylabel("Jumlah Transaksi")
plt.show()
```

```
plt.figure(figsize=(10, 5))
sns.countplot(y="Product line", data=df, palette="viridis", order=df["Product line"].value_counts().index)
plt.title("Kategori Produk Terlaris", fontsize=14, fontweight="bold")
plt.xlabel("Jumlah Terjual", fontsize=12)
plt.ylabel("Kategori Produk", fontsize=12)
plt.show()
```

```
plt.figure(figsize=(8, 5))
df["Product line"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("Set2"))
plt.title("Proporsi Kategori Produk", fontsize=14, fontweight="bold")
plt.ylabel("")
plt.show()
```

Cabang dengan transaksi tertinggi memiliki basis pelanggan yang lebih besar atau strategi pemasaran yang lebih efektif, jika ada cabang dengan transaksi jauh lebih rendah, perlu dievaluasi faktor-faktor penyebabnya. Tindakan yang dilakukan yaitu meningkatkan strategi promosi atau diskon pada cabang dengan transaksi rendah atau dengan menganalisis lokasi dan perilaku pelanggan di setiap cabang.

2. Kategori produk dengan penjualan tertinggi menunjukkan peminat pelanggan yang tinggi, produk dengan penjualan rendah mungkin memerlukan strategi pemasaran atau inovasi baru dan analisis mengapa kategori tertentu kurang diminati pelanggan.

3. Kategori produk yang mendominasi menunjukkan minat belanja pelanggan. Toko perlu menyesuaikan stok dan strategi pemasaran berdasarkan kategori yang paling laris dan selalu mengevaluasi apakah kategori dengan persentase kecil masih layak dijual atau memerlukan inovasi.
